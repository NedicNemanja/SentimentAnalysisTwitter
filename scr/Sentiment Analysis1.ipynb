```python
import pandas as pd
dataset_train = pd.read_csv('twitter11111111111.tsv', encoding='utf8',delimiter='\t');
dataset_test = pd.read_csv("../twitter_data/test2017.tsv", encoding='utf8',delimiter='\t');
dataset_tr = pd.read_csv("../twitter_data/train2017.tsv", encoding='utf8',delimiter='\t');

corpus=[]
for i in range(dataset_train.shape[0]):
    tweet = dataset_train.iloc[i, 0]
    corpus.append(tweet)
    
test_tweets=[]   
for i in range(dataset_test.shape[0]):
    tweet = dataset_test.iloc[i, 3]
    test_tweets.append(tweet) 

#csv_output = []
#file_path="../twitter_data/train2017.tsv"
#with open(file_path, 'r',encoding='utf8',errors="ignore") as f_input:
 #   csv_input = csv.reader(f_input, delimiter='\t', skipinitialspace=True)
  #  for cols in csv_input:
  #      csv_output.append(cols[2])    
#print(csv_output)

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words=nltk_stopwords ,max_features=1000)

#print( vectorizer.fit_transform(corpus).todense() )
#print( vectorizer.vocabulary_)   
X = vectorizer.fit_transform(corpus).toarray()
y = dataset_tr.iloc[:, 2]
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X, y)
X_test= vectorizer.fit_transform(test_tweets).toarray()

y_pred = classifier.predict(X_test)
```
